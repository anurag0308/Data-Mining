```python
def candidate_set(mydata):
    A= []
    B= []
    for column in mydata:
        A.append(column.split(','))
    for i in range(len(A)):
        for j in range(len(A[i])):
            if (A[i][j] not in B):
                B.append(A[i][j])
    return A,B
    
    
    
```
